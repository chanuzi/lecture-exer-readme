```python
"""
Lists
"""

primes=[2,3,5,7,11,2]

print(primes)
#파이선의 리스트는 중복을 허용

empty_list=[]
print(empty_list)
```

```python
"""
Multiple different data types
"""
mixed = ['jenny',1,1.4,True]

print(mixed)
list_of_lists=[['a',1],['b',2]]
print(list_of_lists)
```

```python
"""
Zero Indexing lists
"""
names=['Roger','Rafael','Andy',"novak"]
print(names[0])#first data
print(names[1])
print(names[2])
print(names[3])
```

```python
"""
Negative List indices
"""

names=['Roger','Rafael','Andy',"novak"]
print(names[-1])#last data  
print(names[-2])
print(names[-3])
print(names[-4])
```

```python
"""
List slicing
"""
hsu_buildings=['만우관','장공관','필헌관','소통관','송암관','늦봄관','장준하통일관']
building_slice=hsu_buildings[1:4]#building_slice또한 하나의 독립적인 개체가된다

print(hsu_buildings)
print(building_slice)

building_slice[0]='생활관'
print(building_slice)
print(hsu_buildings[:5])
```

```python
 """
 2-dimensional Lists
 """
restaurants=[['해우리','한식','고s'],['짜장명가','중식','중화비빔밥'],['찌개동아리','한식','제육전골'],['맘스터치','양식','싸이버거']]
print(restaurants[1][2])
restaurants[0][2]='해우라면'
print(restaurants[0])
```

```python
"""
Adding lists together
"""
items_one=['cake','cookie','bread']
items_two=['biscuit','tart']

total_items=items_one+items_two
print(total_items)
```

```python
"""
Useful Methods for Lists
"""
#len()
knapsack=[2,4,3,7,10]
size=len(knapsack)
print(size)

#count()
backpack=['pencil','pen','notebook','textbook',
         'pen','highlighter','pen']
numpen=backpack.count('pen')
print(numpen) 

#append()
orders=['daisies','periwinkle']
orders.append('tulips')
print(orders)

#insert()
store_line=['karla','maxium','martin','isabella']
store_line.insert(2,"vikor")
print(store_line)

#pop()
sw_subjects=['C programming','Calculus','Data Engineering','Data Structures',
            "Machine Learnings"]
sw_subjects.pop(2)

print(sw_subjects)

sw_subjects.pop()
print(sw_subjects)

#remove()
sw_subjects.remove('Calculus')
print(sw_subjects)
# sw_subjects.remove('Algorithms') 리스트에 없는걸 지우면 value에러

#sort()
numbers=[4,2,1,3]
numbers.sort()
print(numbers)

#숫자뿐만아니라 문자까지 정렬가능
hsu_buildings.sort()
print(hsu_buildings)

#sorted()
#새로운 리스트에 따로 생성후 저장가능

sorted_numbers=sorted(numbers)
print(sorted_numbers)
```
