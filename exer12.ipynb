```python
"""
Date pandas
"""

import pandas as pd

date=pd.to_datetime("27th of October,2021")
print(date)

date.strftime('%A')
```

```python
k=date+pd.to_timedelta(np.arange(7))
print(k)
```

```python
ind=pd.DatetimeIndex(['2021-10-27','2021-10-28','2021-10-29','2021-10-30',
                 '2021-10-31','2021-11-01','2021-11-02'])
print(ind)

ser=pd.Series([0,1,2,3,4,5,6],index=ind)
print(ser)

print(ser['2021-10-27':'2021-10-31'])
print(ser['2021-10'])
```

```python
dates=pd.to_datetime([datetime(2015,7,3),
                     "4th of July,2015",
                     '2015-jul-06',
                     '07-07-2015',
                     '20150708'])
print(dates)
```

```python
dates.to_period('D')
dates.to_period('w')
```
