```python

np.random.seed(0)
x=np.random.randint(10,size=(3,4))
print(x)

print(x<6)
print(np.less(x,6))
```

```python
"""
working with boolean arrays
"""

print(x)
print(np.count_nonzero(x<6))
print(np.sum(x<6))

print(np.any(x>8))
print(np.all(x>1))
```

```python
#masking operations

print(x>5)

print(np.std(x[x<5]))
```

```python
"""
Fancy Indexing
"""

np.random.seed(1)
x=np.random.randint(100,size=10)
print(x)

print(x[2],x[4],x[6])

ind=[2,4,6]
print(x[ind])
```

```python
"""
Fancy Indexing n-d arrays
"""

np.random.seed(2)
x=np.random.randint(10,size=(3,5))
print(x)

row=[0,1,2]
col=[1,3,2]
print(x[row,col])
```

```python
"""
Fancy Indexing + broadcasting
"""

print(x)
row=np.array([0,1,2])
col=np.array([1,3,2])
print(x[row,col])

print(x[row[:,np.newaxis],col])
```

```python
"""
Combined Indexing
"""
x=np.arange(1,13).reshape(3,4)
print(x)

print(x[2,[2,0,1]])
print(x[1:,[2,0,1]])
```

```python
"""
Example: Selecting Random Points
"""

mean=[0,0]
cov=[[1,2],[2,5]]

rand=np.random.RandomState(1)
x=rand.multivariate_normal(mean,cov,100)
print(x)
print(x.shape)
```

```python
plt.scatter(x[:,0],x[:,1]);
```

```python
ind = np.random.choice(x.shape[0],20,replace=False)
#100개중 20개를 무작위로 뽑겠다
print(ind)
```

```python
selected=x[ind]
print(selected)
print(selected.shape)
```

```python
plt.scatter(x[:,0],x[:,1],alpha=0.3);#원본
plt.scatter(selected[:,0],selected[:,1],facecolor='red',s=50);
```
