```python
"""
for Loops
"""

nums=[1,2,3,4,5]
for num in nums:  #for variable in list: 
    print(num)
print('\n')    

    
#indentationerror 들여쓰기잘못했을때뜨는에러

for i in range(4):
    print(i)
```

```python
"""
continue keyword
"""

nums=[1,2,-1,4,-5,5,2,-9]
for num in nums:
    #if num<0  we have to skip it
    if num<0:
        continue
    print(num)
```

```python
"""
break keyword
"""
game_plays=['stage1','stage2','stage3','gameover','stage4','stage5']

for play in game_plays:
    if play=='gameover':
        print("Game is over...")
        break
    print(play+"is complete.")
```

```python
"""
List Comprehension
"""
one_to_ten=[i+1 for i in range(10)]
print(one_to_ten)
    
    
```

```python
"""
List Comprehension including conditions
"""

hotspurs=[['손흠민','a+'],['해리 케인','f'],
         ['탕기 은돔벨레','f'],['델리 알리','b'],
         ['루카스 모우라','b+'],['이메르송','d']]
poors=[player[0] for player in hotspurs if player[1]=='f']

print(poors)
```

```python
"""
nested loops
"""
#구구단 프로그램 (1~9)
one_to_nine=[i+1 for i in range(9)]

for i in one_to_nine:
    for j in one_to_nine:
        print("%d x %d = %d" %(i,j,i*j))
```
