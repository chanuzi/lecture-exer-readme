```python
"""
Function vs Method
"""
#function
def my_function(x):
    return x+1

print(my_function(1))

y=3
print(my_function(y))
    
# method:클래스 내에 정의된 함수
x=1

class my_class:
    #@staticmethod
    def my_method(self):
        return x+1

cls=my_class()

print(cls.my_method())

#만약에 클래스밖에서 변수를 정의하는게 아닌 안에서 받아서
#하고싶으면 클래스안에서 @staticmethod를 사용하고 본문에서 받을수있다
#그렇게하면 my_method 의 매개변수에 함수의내용에사용되는 변수와 같은변수를 사용해야함
```

```python
"""
Return keyword
"""
#leap year:윤년(2월29일까지있는 년도)

def check_leap_year(year):
    if year%4==0:
        return str(year)+" is a leap year."
    else:
        return str(year)+" is not a leap year"
result=check_leap_year(2021)  #함수에 2021인자를 넣어서 호출한값을 result에 할당
print(result)
```

```python
"""
Returning Multiple Values
"""

def square_point(x,y,z): 
    x_squared=x**2
    y_squared=y**2
    z_squared=z**2
    return x_squared,y_squared,z_squared
    
three_squared,four_squared,five_squared=square_point(3,4,5)
print(three_squared,four_squared,five_squared)
```

```python
"""
Gloval variables
"""

a="hello"

def prints_a():
    print(a)
    
prints_a()
```

```python
"""
Local variables
"""

def prints_b():
    b="hello"
    #print(b)

prints_b()
print(b)

#지역변수이므로 함수밖에서는 b라는 변수 사용불가하다
#실행되려면 함수안에 프린트문을 쓰고 호출만해야한다
```

```python
"""
Keyword Arguments and Default Values
"""
#부피구하기
#함수의 파라미터에 default값 생성가능 
#본문에서 입력시 daefault값대신 입력값사용
#def findvolume(height,width,depth=3):이런식
def findvolume(height,width,depth=3):
    print("height= "+str(height))
    print("width= "+str(width))
    print("depth= "+str(depth))
    return height*width*depth

print(findvolume(1,2,3))
print(findvolume(width=4,height=3,depth=5))
print(findvolume(width=4,height=3))#안되는건데 위에default값 정해줘서 됨
```
