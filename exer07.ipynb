```python
import numpy as np
```

```python
"""
Numpy Array Attriboutes
"""
print(np.arange(3))
print(np.arange(3,9))
print(np.arange(3,9,2))

```

```python
"""
Random number generator
"""
np.random.seed(0)#이걸 안쓰면 엔터칠때마다 난수가 바뀜
arr1=np.random.randint(10,size=6)
print(arr1)
arr2=np.random.randint(10,size=(3,4))
print(arr2)

arr3=np.random.randint(10,size=(3,4,5))
print(arr3)

```

```python
#Attributes
print(arr2)

print("arr2.ndim=",arr2.ndim)
print("arr2.shape=",arr2.shape)# 제일 많이 사용
print("arr2.siz2=",arr2.size) #**
print("arr2.dtype=",arr2.dtype) #**
print("arr2.itemsize=",arr2.itemsize)
print("arr2.nbytes=",arr2.nbytes)
```

```python
"""
Slicing
"""
arr=np.arange(10)
print(arr)
print(arr[3:7])
print(arr[:5])
print(arr[5:])
print(arr[::3])
print(arr[2:8:2])



```

```python
#Multi-dimensional subarrays
np.random.seed(0)
arr=np.random.randint(10,size=(6,8))
print(arr)

arr_slice=arr[3:5,4:7]
print(arr_slice)
```

```python
"""
#Multi-dimensional subarrays exercise
"""

np.random.seed(0)
arr=np.random.randint(10,size=(5,7))
print(arr)

print(arr[:2,:3])
print(arr[:3,::2])
```

```python
"""
Subarrays as no-copy views
"""
np.random.seed(0)
arr=np.random.randint(10,size=(3,4))
print(arr,'\n')

arr_sub=arr[:2,:2]
print(arr_sub,'\n')

arr_sub[0,0]=99
print(arr)
```

```python
#만약에 슬라이싱부분을 변경해도 원본을 잃기 싫다
#.copy()
np.random.seed(0)
arr=np.random.randint(10,size=(3,4))
print(arr)

arr_sub=arr[:2,:2].copy()
print(arr_sub,'\n')

arr_sub[0,0]=99
print(arr)
```

```python
"""
Reshaping of Arrays
"""
arr=np.arange(1,10)
print(arr,'\n')

grid=arr.reshape(3,3)
print(grid)
```

```python
#np.newaxis

arr= np.array([1,2,3])
print(arr)
print("arr.ndim=",arr.ndim)
print("arr.shape= ",arr.shape)

arr_new=arr[np.newaxis,:]
print(arr_new)
print("arr_new.ndim=",arr_new.ndim)
print("arr_new,shape= ",arr_new.shape)
```

```python
"""
Arrat Concatenation
"""

arr_1=np.array([1,2,3])
arr_2=np.array([4,5,6])
arr_3=np.array([7,8,9])

arr_total=np.concatenate([arr_1,arr_2,arr_3])
print(arr_total)

arr_total=np.vstack([arr_1,arr_2,arr_3])
print(arr_total)
```

```python
"""
Array Splitting
"""

arr=[1,2,3,99,99,3,2,1]
print(arr,'\n')

arr_1,arr_2,arr_3=np.split(arr,[3,5])
print(arr_1,arr_2,arr_3)
```
