```python
"""
Conditional statements(if/elif/else)
"""

puppy="maltepoo"
is_yes=="false"

if puppy=="maltese":
    print("말티즈")
    is_yes="true"

elif puppy=="puddle":
    print("푸들")
    is_yes="true"

elif puppy=="shitzu":
    print("시츄")
    is_yes="true"
    
else:
    print("아 말티푸구나")    
    is_yes="true"
    
print(is_yes)    
    
```
