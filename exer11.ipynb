```python
import numpy as np
import pandas as pd
```

```python
"""
numpy.concatenate()
"""
x=[1,2,3]
y=[4,5,6]
z=[7,8,9]

np.concatenate([x,y,z])
#vstack():concatenate certically
#hstack():concatenate horizontally

np.vstack([x,y,z])
```

```python
"""
Pandas.concat()
"""
ser1=pd.Series(['a','b','c'],index=[1,2,3])
ser2=pd.Series(['d','e','f'],index=[4,5,6])
print(pd.concat([ser1,ser2]))
print(pd.concat([ser2,ser1]))
```

```python
def make_df(cols,ind):
    data={c:[str(c)+str(i)for i in ind]
         for c in cols}
    return pd.DataFrame(data,ind)

print(make_df('abc',[1,2]))
```

```python
df1=make_df('AB',[1,2])
df2=make_df('AB',[3,4])
print(df1,'\n')
print(df2,'\n')

print(pd.concat([df1,df2]))
```

```python
df3=make_df('AB',[0,1])
df4=make_df('CD',[0,1])
print(df3,'\n')
print(df4,'\n')

print(pd.concat([df3,df4],axis=1))
print(pd.concat([df3,df4],axis=0))
```

```python
"""
concat() Features
"""
df1=make_df('AB',[1,2])
df2=make_df('AB',[3,4])

df2.index=df1.index

print(df1,'\n')
print(df2,'\n')

print(pd.concat([df1,df2]))

```

```python
try:
    pd.concat([df1,df2],verify_integrity=True)
    
except ValueError as e:
    print("ValueError: ",e)
```

```python
#인덱스무시
print(pd.concat([df1,df2], ignore_index=True))
```

```python
"""
concat():joinoption
"""
x=make_df('ABC',[1,2])
y=make_df('BCD',[3,4])

print(x,'\n')
print(y,'\n')

print(pd.concat([x,y],join='outer'))
print(pd.concat([x,y],join='inner'))
```

```python
"""
Join operations
"""

df1=pd.DataFrame({'employee':['Bob','Jake','Lisa','Sue'],
                 'group':['Accounting','Engineering','Engineering','HR']})
df2=pd.DataFrame({'employee':['Lisa','Bob','Jake','Sue'],
                 'hire_date':['2004','2008','2012','2014']})

print(df1,'\n')
print(df2,'\n')

```

```python
#일대일 조인의 merge
df3=pd.merge(df1,df2)
print(df3)
```

```python
#다대일 조언의 merge

df4=pd.DataFrame({'group':['Accounting','Engineering','HR'],
                 'supervisor':['Carly','Guido','Steve']})
print(df3,'\n')
print(df4,'\n')

pd.merge(df3,df4) #df3=lefy df4=right
```

```python
df5=pd.DataFrame({'group':['Accounting','Accounting','Engineering','Engineering','HR','HR'],
                 'skills':['math','spreadsheets','coding','linux','spreadsheets','organization']})
print(df1)
print(df5)
print(pd.merge(df1,df5))
```

```python
"""
Specifying the merge key
"""

print(df1,'\n')
print(df2,'\n')

df1['age']=pd.Series([20,42,37,25])
df2['age']=pd.Series([37,20,42,25])
print(df1,'\n')
print(df2,'\n')

print(pd.merge(df1,df2,on='employee'))
print(pd.merge(df1,df2,on='age'))
```

```python
# left on right on

df3=pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
                 'salary':[70000,80000,120000,90000]})

print(df1,'\n')
print(df3,'\n')

print(pd.merge(df1,df3,left_on='employee',right_on='name'))
```

```python
#left_index,right_index

print(df1,'\n')
print(df2,'\n')

df1a=df1.set_index('employee')
df2a=df2.set_index('employee')
print(df1a,'\n')
print(df2a,'\n')

print(pd.merge(df1a,df2a,left_index=True,right_index=True))
```
